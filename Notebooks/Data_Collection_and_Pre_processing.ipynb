```python
# ===== Import libraries ===== #
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

```python
# Search for target protein (Chembl database)
target = new_client.target
target_query = target.search("coronavirus")
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
# 3 single proteins
# Select and retrieve bioactivity data for SARS Cov 3C-like proteinase (6th entry)
selected_target = targets.target_chembl_id[6]
selected_target # unique id for target protein
```

```python
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

```python
df = pd.DataFrame.from_dict(res)
```

```python
df.head(3)
```

```python
df.standard_type.unique() # ensure = same bioactivity unit
```

```python
# Need to look at a low standard value (low C of drug & high Pot)
df.to_csv("bioactivity_data.csv", index=False)
```

```python
# Data cleaning
```

```python
df2 = df[df.standard_value.notna()]
df2 # no missing data
```

```python
# Data pre-processing for ML classifier (active, innactive and intermediate) based on IC50 val
```

```python
bioactivity_class = []
for i in df2.standard_value:
  if float(i) >= 10000:
    bioactivity_class.append("inactive")
  elif float(i) <= 1000:
    bioactivity_class.append("active")
  else:
    bioactivity_class.append("intermediate")
```

```python
# Each compound = described by a molecule chembl id
# If chembl id = same across >1 row = keep only 1 - eliminate redundancy
```

```python
df2.molecule_chembl_id
```

```python
mol_cid = []
for i in df2.molecule_chembl_id:
  mol_cid.append(i)
```

```python
mol_cid
```

```python
cannonical_smiles = []
for i in df2.canonical_smiles:
  cannonical_smiles.append(i)
```

```python
standard_value = []
for i in df2.standard_value:
  standard_value.append(i)
```

```python
selection = ["molecule_chembl_id", "canonical_smiles", "standard_value"]
df3 = df2[selection]
```

```python
df3
```

```python
pd.concat([df3, pd.Series(bioactivity_class)], axis=1)
```

```python
# Create CSV file for pre-processed bioactivity data set
```

```python
df3.to_csv("bioactivity_preprocessed_data.csv", index=False)
```
